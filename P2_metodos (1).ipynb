# Modelo **(SEIQR)** ü¶†

O modelo SIR foi proposto em 1927 por W.O. Kermack em conjunto com A.G. McKendrick, nesse modelo √© considerada a exist√™ncia de uma popula√ß√£o fixa em que cada individuo pode ser classificado em apenas uma das. 3 classes: S (sens√≠vel), I (infectado) ou R (removido), em que S s√£o os indiv√≠duos suscet√≠veis a contrair a doen√ßa, I os que possuem a doen√ßa e s√£o capazes de transmitir e R os que j√° foram infectados e agora s√£o imunes, ou seja, n√£o podem ser infectados novamente ou infectar outros. 

Nessa pesquisa, a base √© o modelo SIR, por√©m ser√£o adicionadas mais duas classes: E (exposto) e Q (quarentenado), em que E s√£o os indiv√≠duos infectados mas que n√£o s√£o capazes de transmitir e Q os indiv√≠duos que testaram positivo para a doen√ßa e se mantiveram em quarentena. Formando assim o modelo SEIQR.

O modelo SEIQR √© composto por cinco equa√ß√µes ordinarias diferenciais (EDOs) que representam o n√∫mero de pessoas em cada classe em determinado momento:

$\dfrac{dS}{dt} = Œõ - ŒºS - Œ≤S\dfrac{I}{S+E+I+R}$

$\dfrac{dE}{dt} = Œ≤S\dfrac{I}{S+E+I+R} - (Œº + Œµ + Œ∑Œª)E $

$\dfrac{dI}{dt} = ŒµE - (Œº+$Œ≥$+Œ±+Œ∑Œª)I$

$\dfrac{dQ}{dt} =Œ∑Œª(E +I)-(Œº+Œ≥ +Œ±)Q$

$\dfrac{dR}{dt} =  Œ≥(Q + I)- ŒºR$

Onde:

Œõ : Taxa de nascimento por unidade de tempo\
Œº : Taxa de mortalidade por pessoa por unidade de tempo\
$Œ±$ : Taxa de fatalidade do COVID-19\
$Œ≤$ : Probabilidade de transmiss√£o da doen√ßa por tempo de contato e por n√∫mero de contato por unidade de tempo\
Œµ : Taxa de progress√£o da exposi√ß√£o da infec√ß√£o por periodo de tempo\
$Œ≥$ : Taxa de recupera√ß√£o de indiv√≠duos infectados do per√≠odo infeccioso\
Œ∑ : Sensibilidade ao teste r√°pido\
Œª : Porcentagem da popula√ß√£o que realizou o teste r√°pido



Iremos considerar que pessoas sens√≠veis possuem a probabilidade de encontrar algu√©m que pode transmitir a doen√ßa de:  \
\
$ \dfrac{I}{S + E + I + R}$ 


Al√©m disso podemos observar que o n√∫mero de pessoas que passam de ses√≠veis para expostas por unidade de tempo √© representado como : \
\
$ Œ≤\dfrac{I}{S + E + I + R}S$ 

Tamb√©m vemos que a taxa em que uma pessoa exposta (E) se torna quarentenada (Q) √© representada como: \
\
$ Œ∑ŒªE$ 

N√≥s tamb√©m assumiremos que N √© o n√∫mero total de pessoas na popula√ß√£o, ou seja:\
\
$N'(t) = S'(t) + E'(t) + I'(t) + Q'(t) + R'(t) = Œõ - Œº(S +E +I +Q+R)- Œ±(I +Q) $


# Visualizando os dados üîç

### Importando as bibliotecas necess√°rias

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

### Definindo os parametros e as condi√ß√µes iniciais

```python
# Initial conditions
S0 = 0.99  
E0 = 0.01
I0 = 0.01  
R0 = 0.00  
Q0 = 0.00
y0 = [S0, E0, I0, R0, Q0]

# Parameters

BR = 0.00005 #Birth rate per unit of time (Œõ)
DR = 0.000205 #Death rate per unit of time (Œº)
PT = 5 #Probability of disease transmition per unit of time (Œ≤)
PR = 0.33 #Progression rate from exposed to infectious (Œµ)
FR = 0.0336 #Fatality rate of covid19 (Œ±)
RR = 0.2 #Recovery rate (Œ≥)
SE = 0.1 #Sensitity of rapid test (Œ∑)
YR = 0.1 #Percentage of people who takes the test (Œª)

# Time vector
t = np.linspace(0, 40, 200)  
```

### Definindo o modelo e solu√ß√£o das equa√ß√µes

```python
def SEIR_model(y, t, BR, DR, PT, PR, FR, RR, SE, YR):
    S, E, I, R, Q = y
    a = I/(S+E+I+R)
    k = SE*YR
    b = DR + RR + FR
    dSdt = BR - (DR*S) - (PT*S*a)
    dEdt = PT*S*a - E*(DR + PR + k)
    dIdt = PR*E - I*(b + k)
    dRdt = RR*(Q+I) - DR*R
    dQdt = k*(E+I) - b*Q
    return [dSdt, dEdt, dIdt, dRdt, dQdt]
```

### Resolvendo as  equa√ß√µes

```python
# Solve the SEIR model equations
solution = odeint(SEIR_model, y0, t, args=(BR, DR, PT, PR, FR, RR, SE, YR))

# Extract results
S, E, I, R, Q = solution.T
```

# Imprimindo resultados üìà

Seguindo a varia√ß√£o de uma constante k = SE*YR 
(sensibilidade ao teste r√°pido * porcentagem de pessoas que realizaram o teste)

Com isso conseguimos observar a varia√ß√£o de pessoas infectadas, em quarentena, suscet√≠veis √† doen√ßa, expostas e removidas, de acordo com o aumento do n√∫mero de testes r√°pidos feitos em uma popula√ß√£o e o aumento da sua eficiencia.

```python
# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(t, I, label='k = 0.01')
solution = odeint(SEIR_model, y0, t, args=(BR, DR, PT, PR, FR, RR, 0.1, 1))
S, E, I, R, Q = solution.T
plt.plot(t, I, label='k = 0.10')
solution = odeint(SEIR_model, y0, t, args=(BR, DR, PT, PR, FR, RR, 0.2, 1))
S, E, I, R, Q = solution.T
plt.plot(t, I, label='k = 0.20')
solution = odeint(SEIR_model, y0, t, args=(BR, DR, PT, PR, FR, RR, 0.5, 1))
S, E, I, R, Q = solution.T
plt.plot(t, I, label='k = 0.50')
solution = odeint(SEIR_model, y0, t, args=(BR, DR, PT, PR, FR, RR, 1, 1))
S, E, I, R, Q = solution.T
plt.plot(t, I, label='k = 1')
plt.xlabel('Time (days)')
plt.ylabel('Proportion of Population')
plt.title('SEIQR Model Simulation of Infected Population')
plt.legend()
plt.grid(True)
plt.show()
```

Assim, √© poss√≠vel definir –∫=Œ∑Œª. Tamb√©m √© importante ressaltar que, uma vez que Œ∑ e Œª s√£o valores unit√°rios, o valor representado por –∫ tamb√©m ser√°. Com esses valores, simulamos a din√¢mica por meio da resolu√ß√£o das EDOs.
